## **Feature Selection**

```python
df = df[['age', 'workclass', 'fnlwgt', 'education.num',
         #'education', 'marital.status','relationship', 
        'occupation', 'race', 'sex',
       'capital.gain', 'capital.loss', 'hours.per.week', 'native.country',
       'income']]
```

```python
df = pd.get_dummies(df,columns=['occupation'],dtype=int)            # Splitting occupation into different columns
```

# **Training the Model**

## **Data Splitting**

```python
from sklearn.model_selection import train_test_split
x = df[['age', 'workclass', 'fnlwgt', 'education.num', 'race', 'sex',
       'capital.gain', 'capital.loss', 'hours.per.week', 'native.country',
        'occupation_?', 'occupation_Adm-clerical',
       'occupation_Armed-Forces', 'occupation_Craft-repair',
       'occupation_Exec-managerial', 'occupation_Farming-fishing',
       'occupation_Handlers-cleaners', 'occupation_Machine-op-inspct',
       'occupation_Other-service', 'occupation_Priv-house-serv',
       'occupation_Prof-specialty', 'occupation_Protective-serv',
       'occupation_Sales', 'occupation_Tech-support',
       'occupation_Transport-moving']]
y = df[['income']]
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)
```

## **Model Fitting**

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

## **Predicting the values**

```python
y_pred = regressor.predict(x_test)
df_pred = pd.DataFrame({'Actual':y_test.squeeze(),'Predicted':y_pred.squeeze()})
df_pred
```

# **Evaluating the Model**

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
print(mae)
print(mse)
```
